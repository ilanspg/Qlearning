<p>The Following code will un an instance of Cartpoke-v0 environment, rendering the environment each step. (If you run this in Python there will be a pop-up window).</p>
<p>The environment's <b>step</b> function returns four values:</p>
<ul>
    <li>observation(object): an environment-specific ject representing your observation of the environment.</li>
    <li>reward(float): amount of reward achieved by the previous action.</li>
    <li>done(boolean): whether it's time to reset the environment again</li>
    <li>info(dict): dignostic information useful for debugging.</li>
</ul>
<p>This is an implementation of "agent-environmen loop". Each timestep, the agent chooses an action, and the environment returns an observation and a reward.</p>
<p>The process gets sarted by calling reset(), which return an initial observation.</p>
<p>This should give a video and output where you should be able to see where the resets happen.</p>

```python
import gym
env = gym.make('CartPole-v0')
for i_episode in range(20):
    observation = env.reset()
    for t in range(100):
        env.render()
        print(observation)
        action = env.action_space.sample()
        observation, reward, done, info = env.step(action)
        if done:
            print("Episode finished after {} timesteps".format(t+1))
            break
env.close()
```
